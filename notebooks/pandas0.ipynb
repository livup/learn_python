#### pd.DataFrame 

```python
columns = ['item_id', 'analysis_category', 'analysis_subcategory', 'is_active', 'name', 'price']
```

```python
items = items_bq[columns].copy()
```

```python
items.sort_values(['price', 'name'], ascending=[False, True], inplace=True)
```

```python
items
```

```python
items['price'] = items['price'] / 100
```

```python
items['price_dolar'] = items['price'] / 5.10
```

```python
items
```

```python
items.drop('price_dolar', axis=1, inplace=True)
```

```python
items
```

### Filtros

```python
items.head(10)
```

```python
mask = (items.price < 120) & (items.price > 1)

items = items[mask]
```

```python
items = items[items.is_active]
```

#### Apply + Lambda 

```python
def check_price_category(price):
    if price >= 20:
        return 'caro'
    else:
        return 'barato'
```

```python
# NÃO É BOA PRÁTICA!
items['categoria_preço'] = items.price.apply(check_price_category)
```

```python
items
```

#### Agregações

```python
items.groupby('analysis_category').price.mean()
```

```python
agg = {
    'price': ['mean', 'std'],
    'item_id': 'count'
}
```

```python
grouped = items.groupby(['analysis_category', 'analysis_subcategory']).agg(agg)
```

```python
grouped
```
