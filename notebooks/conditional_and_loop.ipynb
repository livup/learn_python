### ..:: Condicional ::..

##### Quando utilizamos?
Os Operadores Condicionais são utilizados para fazer as comparações dos valores que são passados e retornam o valor Verdadeiro ou Falso.


```python
# If e operadores
if 1 == 1:
    print('OK')
    print('hello')
```

#### Operadores
- if
- else
- elif
- ternaria - short hand if else

```python
# function checagem de temperatura do kadu

temp = 23

def verificarTemperaturaKadu(temp=23):
    if temp < 15:
        print ('frio')
    elif temp < 25:
        print('temperatura ideal')
    elif temp < 30:
        print('clima de praia')
    else:
        print('Muito quente')

verificarTemperaturaKadu(temp=55)
```

```python
# ternaria -> short hand if
a = 1
b = 1
print('Sao iguais') if a == b else print('sao diferentes')
```

##### Operadores lógicos

#### AND
Utilizado para verificar quando **duas condições** são verdadeiras
#### OR
Utilizado para verificar quando **uma das** condições são verdadeiras

```python
# Testando a condicao and (ambas condicoes precisam ser verdade)
if 1==1 and 2!=3:
    print('ok')
else:
    print('not ok')
```

```python
# Testando a condicao or ( Apenas uma codicao precisa ser verdade)
if 1==2 or 2==3:
    print('ok')
else:
    print('not ok')
```

### Exemplo

Estamos passando um periodo critico e precisamos criar uma forma escalavel de passar sempre a informação para o nivel abaixo da hierarquia, sendo a hierarquia:

> - Presidente
> - Governador
> - Prefeito
> - Vereadores
> - População

```python
nivel_emergencia = 0

def acionaPresidente():
    print('Presidente acionado, onde pais <> BR ')
    acionaGovernador()

def acionaGovernador():
    print('Governador acionado! ')
    acionaPrefeito()
    
    
def acionaPrefeito():
    print('Prefeito acionado! ')
    print('Vereadores acionados')
    print('População avisada! ')


def avisaCasoEmergencia(nivel_emergencia=0):
    if nivel_emergencia == 0:
        print ('Ambiente tranquilo!')
    elif nivel_emergencia == 1:
        acionaPrefeito()
    elif nivel_emergencia == 2:
        acionaGovernador()
    else:
        acionaPresidente()

avisaCasoEmergencia(10)
    
```

### ..:: Loop ::..

##### Quando utilizamos?
Precisamos executar várias vezes seguidas o mesmo trecho de código. 
Utilzamos um loop (ou laço de repetição) que permite executar o mesmo bloco de código enquanto uma condição é atendida.

Em python, possuimos dois tipos de Loop:
> - for
> - while

### while

- Podemos executar um trecho de código enquanto uma afirmação for verdadeira

```python
# Loop basico de 0 a 3
i = 0
while i < 3:
    print(i)
    i = i +1
```

- Filtrar uma lista com **while**

```python
produtos = ['frango com costra', 'papinha de brocolis', 'hamburguer de fraldinha', 'papinha de banana']
# ['frango', 'com', 'crosta']
indice = 0
papinhas = []


def avisaTenhoPapinha(nome_da_papinha):
    print(f'Habemus +1 papinha: {nome_da_papinha}')

#comando de repetição while
while indice < len(produtos):
    
    
    if 'papinha' in produtos[indice]:
        papinhas.append(produtos[indice])
        avisaTenhoPapinha(produtos[indice])
    indice += 1
    
print(f'\nTemos um total de {len(papinhas)} papinhas:' )
print(papinhas)
```

```python
# voce deseja continuar ?
indice = 0
while True:
    
    print(produtos[indice])
    indice += 1

    if indice >= len(produtos):
        break
```

### for

- Um for loop é utlizado para iterar sob uma sequencia: (list, a tuple, a dictionary, a set, or a string).

```python
#Iterar sobre for em range de 0 - 5
for i in range(5):
    print(i)
```

```python
# Percorrendo os valores da lista de produtos
#Python é semantico:para cada produto estou printando o produto
#Estou falando ==o que faço
for produto in produtos:
    print(f'Produto: {produto}')
```

```python
# percorrer todos os ingredients da tupla
tupla_ingredientes = ( 'sal', 'acucar', 'agua' )
for ing in tupla_ingredientes:
    print(ing)
```

##### List comprehension 

```python
# Criar uma produtos_2 com os mesmo valores de produtos

produtos_2 = [pro for pro in produtos ]
print(produtos_2)
```

- Adicionando um filtro

```python
# Criar um filtro para trazer somente as papinhas
papinhas_2 = [pro for pro in produtos if pro.split(' ')[0] == 'papinha']
print(papinhas_2)
```

- Alterando o objeto para nova lista com filtro

```python
# Alterando a lista para retornar apenas os sabores das papinhas
sabores_papinhas = [prod.split(' ')[-1] for prod in produtos if 'papinha' in prod]
print(sabores_papinhas)
```

##### Dict comprehension 

```python
# Criar um dicionario com chave: papinha e valor: tamanho da string da papinha
papinhas_2 = {item: len(item) for item in produtos}
print(papinhas_2)
```

```python
# Consultando o valor de um dictionary
print(papinhas_2.get('frango com costra'))
```
